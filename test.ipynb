```python
import pandas as pd
import numpy as np

import pyperclip
```

```python
def read_clipboard_df():
    return pd.read_clipboard(sep="\t", index_col=0)

df = read_clipboard_df()
```

```python
df
```

```python

def read_clipboard_pool(all_df=None):
    if all_df is None:
        all_df = pd.read_clipboard(sep="\t", index_col=0)
    players = all_df.index
    df_left = all_df.loc[:, np.arange(len(df.columns))%2==0]
    df_up = all_df.loc[:, np.arange(len(df.columns))%2==1]
    combos = np.identity(len(players))
    lastpicked = []
    decays = [0.01, 0.1, 0.25, 0.5]
    while True:
        probs = 2 ** (len(players) - combos.sum(1)) - 1
        for dec, lastpick in zip(decays, lastpicked):
            for pick in lastpick:
                probs[pick] *= dec
        pick1 = np.random.choice(np.arange(len(players)), p=probs / probs.sum())
        probs2 = probs * (1 - combos[pick1])
        pick2 = np.random.choice(np.arange(len(players)), p=probs2 / probs2.sum())
        combos[pick1, pick2] = 1
        combos[pick2, pick1] = 1
        lastpicked = [[pick1, pick2]] + lastpicked
        if (1 - combos).sum() == 0:
            break
    order = list(reversed(lastpicked))
    final_string = ""
    for match in order:
        match_players = players[match]
        scores = [df_left.iloc[match[0], match[1]], df_up.iloc[match[0], match[1]]]
        if scores[1] > scores[0]:
            match_players = list(reversed(match_players))
        scores = list(sorted(scores, reverse=True))
        final_string += f"{match_players[0]}\t{scores[0]:.0f}\t{scores[1]:.0f}\t{match_players[1]}\n"
    pyperclip.copy(final_string)

read_clipboard_pool(df)
```
